```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
```

```python
ARQUIVO = r'C:\Users\eduarda.castilho\Documents\UC2_aulas\aula05_UC2\03.BaseDPEvolucaoMensalCisp.csv'
df = pd.read_csv(ARQUIVO, sep = ';', encoding ='iso-8859-1')
```

1. **Quantidade de veÃ­culos recuperados por municÃ­pio**

```python
df_munic = df['munic']
df_munic.head()

df_recuperacao_veiculos = df['recuperacao_veiculos']
df_recuperacao_veiculos.head()
```

```python
df_recuperacao_veiculos_munic = df[['munic','recuperacao_veiculos']]

df_recuperacao_veiculos_munic= (
    df_recuperacao_veiculos_munic.groupby('munic')
    .sum().reset_index().sort_values(by = 'recuperacao_veiculos', ascending=False)
)

df_recuperacao_veiculos_munic.head(20)
```

```python
plt.plot(df_recuperacao_veiculos_munic['munic'],df_recuperacao_veiculos_munic['recuperacao_veiculos'],
         marker='o', color='blue')
df_recuperacao_veiculos_munic = df_recuperacao_veiculos_munic.iloc[0:10]
plt.bar(df_recuperacao_veiculos_munic['munic'], df_recuperacao_veiculos_munic['recuperacao_veiculos'])
plt.ylabel('NÃºmero de veÃ­culos recuperados')
plt.title('RecuperaÃ§Ã£o de veÃ­culos por municÃ­pio')
plt.xticks(rotation=90)
plt.show()
```

```python

df_roubo_veiculos_munic = df[['munic','roubo_veiculo']]

df_roubo_veiculos_munic= (
    df_roubo_veiculos_munic.groupby('munic')
    .sum().reset_index().sort_values(by = 'roubo_veiculo', ascending=False)
)


df_roubo_veiculos_munic = df_roubo_veiculos_munic.iloc[0:10]
plt.plot(df_roubo_veiculos_munic['munic'],df_roubo_veiculos_munic['roubo_veiculo'],
         marker='o', color='blue')
plt.bar(df_roubo_veiculos_munic['munic'], df_roubo_veiculos_munic['roubo_veiculo'])
plt.ylabel('NÃºmero de veÃ­culos roubados')
plt.title('NÃºmero de veÃ­culos roubados por municÃ­pio')
plt.xticks(rotation=90)
plt.show()
```

```python
df_ano = df['ano']
df_ano.head()

df_recuperacao_veiculos = df['recuperacao_veiculos']
df_recuperacao_veiculos.head()
```

```python
df_recuperacao_veiculos_ano = df[['ano','recuperacao_veiculos']]

df_recuperacao_veiculos_ano= (
    df_recuperacao_veiculos_ano.groupby('ano')
    .sum().reset_index().sort_values(by = 'recuperacao_veiculos', ascending=False)
)

df_recuperacao_veiculos_ano.head(20)
```
